```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
daily_weather = {
    2021: pd.read_csv("./assets/weather/en_climate_daily_MB_502S001_2021_P1D.csv"),
    2022: pd.read_csv("./assets/weather/en_climate_daily_MB_502S001_2022_P1D.csv"),
    2023: pd.read_csv("./assets/weather/en_climate_daily_MB_502S001_2023_P1D.csv")
}

```

```python
avg_monthly_temp = {}
avg_monthly_percip = {}
avg_monthly_snow_on_ground = {}

for year in [2021, 2022, 2023]:
    for month in range(1, 13):
        # calculate the mean for each month of the year
        avg_monthly_temp[f"{year}-{month}"] = daily_weather[year][daily_weather[year]['Month'] == month]['Mean Temp (C)'].mean()
        avg_monthly_percip[f"{year}-{month}"] = daily_weather[year][daily_weather[year]['Month'] == month]['Total Precip (mm)'].mean()
        avg_monthly_snow_on_ground[f"{year}-{month}"] = daily_weather[year][daily_weather[year]['Month'] == month]['Snow on Grnd (cm)'].mean()



avg_monthly_temp = pd.Series(avg_monthly_temp)[0:25]
avg_monthly_temp.plot(kind="bar")
plt.show()
avg_monthly_percip = pd.Series(avg_monthly_percip)[0:25]
avg_monthly_percip.plot(kind="bar")
plt.show()
avg_monthly_snow_on_ground = pd.Series(avg_monthly_snow_on_ground)[0:25]
avg_monthly_snow_on_ground.plot(kind="bar")
plt.show()
```
